This notebook is a starting point for exploring and visualizing commodity calendar spreads.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dummy data to simulate calendar spread
spread_data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Spread ($)': [0.15, 0.12, 0.18, 0.10, 0.20]
}

df = pd.DataFrame(spread_data)
df.plot(kind='bar', x='Month', y='Spread ($)', title='Simulated Calendar Spread')
plt.ylabel('Price Difference ($)')
plt.tight_layout()
plt.show()
```
